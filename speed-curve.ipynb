```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# %matplotlib inline
%config InlineBackend.figure_format ='svg'
```

```python
angV25=[]
angV25sum=1000
angV25move=0
Flag=True
for i in range(0,angV25sum+1):
    data = pd.read_csv(r'E:\test\speedangV\25\P_'+str(i)+'.csv', usecols=[0,1,2,3])
    if np.size(data):
        uniques = np.unique(data, axis=0)
        x=np.mean(uniques[:,0],axis=0)
        y=np.mean(uniques[:,1],axis=0)
        z=np.mean(uniques[:,2],axis=0)
        v=np.mean(uniques[:,3],axis=0)
        angV25.append([x,y,z,v])
        if Flag:
            angV25move=i
            Flag=False
    else:
        angV25.append([0,0,0,0])
```

```python
angV50=[]
angV50sum=1000
angV50move=0
Flag=True
for i in range(0,angV50sum+1):
    data = pd.read_csv(r'E:\test\speedangV\50\Bunker37_'+str(i)+'.csv', usecols=[0,1,2,3])
    if np.size(data):
        uniques = np.unique(data, axis=0)
        x=np.mean(uniques[:,0],axis=0)
        y=np.mean(uniques[:,1],axis=0)
        z=np.mean(uniques[:,2],axis=0)
        v=np.mean(uniques[:,3],axis=0)
        angV50.append([x,y,z,v])
        if Flag:
            angV50move=i
            Flag=False
    else:
        angV50.append([0,0,0,0])
```

```python
angV100=[]
angV100sum=1791
angV100move=0
Flag=True
for i in range(0,angV100sum+1):
    data = pd.read_csv(r'E:\test\speedangV\100\P_'+str(i)+'.csv', usecols=[0,1,2,3])
    if np.size(data):
        uniques = np.unique(data, axis=0)
        x=np.mean(uniques[:,0],axis=0)
        y=np.mean(uniques[:,1],axis=0)
        z=np.mean(uniques[:,2],axis=0)
        v=np.mean(uniques[:,3],axis=0)
        angV100.append([x,y,z,v])
        if Flag:
            angV100move=i
            Flag=False
    else:
        angV100.append([0,0,0,0])
```

```python
X_angV25=[i*1000*5e-6*0.3 for i in range(0,angV25sum+1)]
X_angV50=[i*1000*5e-6*0.3 for i in range(0,angV50sum+1)]
X_angV100=[i*1000*2e-6*0.3 for i in range(0,angV100sum+1)]
```

```python
N=1
X=3
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['font.sans-serif'] = ['Times New Roman']

plt.plot(X_angV25[::N], [j[X] for j in angV25[::N]], color='C0',label='angV = 25.0')
plt.plot(X_angV50[::N], [j[X] for j in angV50[::N]], color='C1',label='angV = 50.0')
plt.plot(X_angV100[::N],[j[X] for j in angV100[::N]], color='C2',label='angV = 100.0')


plt.xlim(xmin = 0);plt.xlim(xmax = 1.5)
plt.ylim(ymin = 0);plt.ylim(ymax = 1.2)
plt.xlabel('Distance (m)')
plt.ylabel('V (m/s)')
plt.legend()
plt.savefig(r'E:\0001.svg')
plt.show()
```
