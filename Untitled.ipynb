```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format ='svg'
```

```python
# 读取Bunker37的速度

speed_z1=[]
for i in range(74,1001):
#     data = pd.read_csv('/home/zhi/Desktop/Bunker37-speed/Bunker37.'+str(i)+'.csv', usecols=[0,1,2,3])
    data = pd.read_csv(r'E:\07-24\Bunker37-speed\Bunker37.'+str(i)+'.csv', usecols=[0,1,2,3])
    uniques = np.unique(data, axis=0)
    z=np.mean(uniques[:,3],axis=0)
    speed_z1.append(z)
```

```python
xx1=[i for i in range(90,1001)]
yy1=speed_z1[90-74:]
f1 = np.polyfit(xx1, yy1, 1)
p1 = np.poly1d(f1)
print(p1)#打印出拟合函数
yval1 = p1(xx1)  #拟合y
```

```python
plt.plot([i for i in range(74,1001)], speed_z1)
# plt.plot(xx1, yval1)
plt.xlim(xmin = 0)
plt.ylim(ymax = 0,ymin=-1)
plt.xlabel('iter')
plt.ylabel('Z_vel')
plt.show()
```
